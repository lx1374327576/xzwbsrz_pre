```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import xgboost as xgb
import csv
from sklearn import preprocessing
from keras.layers import LSTM, Dense
from keras.models import Sequential
from keras.layers.core import Masking
from keras.layers.embeddings import Embedding
df1 = pd.read_csv("data/train_preliminary/user.csv")
df2 = pd.read_csv("data/train_preliminary/ad.csv")
df3 = pd.read_csv("data/train_preliminary/click_log.csv")
print('read success')
```

```python
df = pd.merge(df3, df2, left_on='creative_id', right_on='creative_id')
order = ['user_id', 'time', 'creative_id', 'click_times', 'ad_id', 'product_category', 'advertiser_id']
df = df[order]
df.sort_values(['user_id', 'time'], inplace=True)
df.reset_index(drop=True, inplace=True)
print('merge success!')
```

```python
x_tmp = np.array(df['creative_id'])
x_tmp = x_tmp.reshape((-1, 1))
x_tmp = np.vstack((x_tmp, np.zeros((9, 1))))
x_tmp = x_tmp.reshape(-1, 10)
x_tmp.shape
```

```python
model = Sequential()
model.add(Embedding(5000000, 64, input_length=x_tmp.shape[1]))
model.compile(loss='MSE', optimizer='adam')
y_tmp = model.predict(x_tmp, verbose=1)
y_tmp = y_tmp.reshape(-1, 64)
```

```python
y_tmp = pd.concat([df['user_id'], pd.DataFrame(y_tmp)], axis=1)
```

```python
y_tmp = y_tmp.iloc[:30082771, :]
y_tmp
```

```python
X = y_tmp.groupby('user_id').sum()
```

```python
X = X.reset_index(drop='True')
X.shape
```

```python
X = preprocessing.MinMaxScaler().fit_transform(X)
Y2_train, Y2_test = df1.iloc[:800000, 2:3], df1.iloc[800000:, 2:3]
X_train, X_test = X[:800000, :], X[800000:, :]
Y2_train = np.array(Y2_train)
model = Sequential()
model.add(Dense(8, input_dim=64, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, kernel_initializer='normal'))
model.compile(loss='MSE', optimizer='adam')
model.fit(X_train, Y2_train, epochs=20)
```

```python
Y1_train, Y1_test = df1.iloc[:800000, 1:2], df1.iloc[800000:, 1:2]
Y1_train = pd.get_dummies(Y1_train['age'])
Y1_train = Y1_train.iloc[:,2]
Y1_train = np.array(Y1_train)
model1 = Sequential()
model1.add(Dense(8, input_dim=64, kernel_initializer='normal', activation='relu'))
model1.add(Dense(1, kernel_initializer='normal'))
model1.compile(loss='MSE', optimizer='adam')
model1.fit(X_train, Y1_train, epochs=20)
```
